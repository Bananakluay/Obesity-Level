Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Import Data

```python
df = pd.read_csv("ObesityDataSet_raw_and_data_sinthetic.csv").copy()
```

Cleaning Data

```python
#check Nulls
df.isnull().sum()
```

```python
# Check Duplicates
print(df.duplicated().sum())
```

```python
# Drop Duplicates 
df.drop_duplicates(inplace=True)
print(df.duplicated().sum())
```

```python
#Round 
behaviours = ["Age", "FCVC", "NCP", "CH2O", "FAF", "TUE"]
for behaviour in behaviours :
    df[behaviour] = df[behaviour].round()
```
